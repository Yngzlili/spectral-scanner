## zigzag

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_zigzag_path(range_xy, step_xy):
    range_x,range_y = range_xy
    step_x,step_y = step_xy
    max_x = range_x//step_x
    max_y = range_y//step_y
    xs= np.linspace(-max_x,max_x,(abs(max_x)*2//step_x+1)) # neg_x -> pos_x
    ys = np.linspace(-max_y,max_y,(abs(max_y)*2//step_y+1)) # neg_y -> pos_y                 
    print(xs.shape,ys.shape)
    nPix = len(xs)*len(ys)
    xyPath = np.zeros((nPix, 2))
    
    ## assume it start from the left corner (+,+)
    iPix = 0
    for y in ys:
        for x in xs:
            if abs(y)%2==0:
                # even row, pos_x->neg_x, againt the xs' sorting
                xyPath[iPix]=[-x,-y]
            else:
                # odd row, neg_x - > pos_x, along the xs' sorting
                xyPath[iPix]=[x,-y]
            iPix = iPix+1
    return xyPath
```

```python
# Parameters
size_xy = (90,88)  # grid size (columns, rows)
step_xy = (2,2)    # step size in each direction

# Generate path
zigzag_path = generate_zigzag_path(size_xy, step_xy)
```
